```python

#Khai báo các thư viện
import numpy as np
import matplotlib.pyplot as plt
#Khai báo các biến input
v_0 = float(input("Hãy nhập giá trị v_o: ")) #Vận tốc thẳng đứng v_0 (m/s)
a = float(input("Hãy nhập giá trị a: ")) #Hằng số gió a (m^-1)

#Phương trình chuyển động:
t = 5
y = v_0 * t # Phương trình chuyển động theo phương thẳng đứng là phương trình chuyển động đều
# v_x = a*y = a * v_o * t => a * v_0 là hằng số gia tốc theo phương ngang <=> a_x = a * v_0
# Khi ấy: phương trình chuyển động theo phương ngang sẽ là phương trình chuyển động nhanh dần đều
a_x = a * v_0
x = (1/2) * a_x * (t**2)

#Phương trình quỹ đạo:
#y = ((v_0 / a) * (2*x)) **(1/2)

#Vẽ quỹ đạo của vật trong khoảng thời gian từ 0 đến 5s
t_values = np. linspace(0, 5, 100) #Gán giá trị t lần lượt từ start = 0 đến end = 5 với bước nhảy = 100
y_values = v_0 * t_values #Tính giá trị y gắn lần lượt với các giá trị t_values
x_values = (1/2) * a_x * (t_values**2) #Tính giá trị x gắn lần lượt với các trị t_values

#Vẽ đồ thị chuyển động
plt.plot(x_values, y_values, label = 'Quỹ đạo khí cầu')
plt.scatter([x], [y], color = 'red', zorder = 5, label = f'Vị trí tại t = {t}s') #Tạo biểu đồ phân tán
plt.xlabel('Quãng đường theo phương ngang (x) (m)')
plt.ylabel('Quãng đường theo phương thẳng đứng (y) (m)')
plt.title('Quỹ đạo chuyển động của khinh khí cầu')
plt.legend()
plt.grid(True)
plt.show()

#In kết quả
print(f'Tại thời điểm t = {t} giây:')
print(f'Độ cao y = {y} m')
print(f'Quãng đường theo phương ngang x = {x} m')

#Xin cảm ơn Beyonce:33333
```
